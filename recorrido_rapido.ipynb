```python
# Funcion generadas al vuelo 

def tabla(n, operador="*", cantidad=5):               
    print(f"Tabla del {n}")
    for i in range(1, cantidad+1):         
        r = eval(f"{n}{operador}{i}")       # JS > r = eval(`${a}${operador}${b}`)
        print(f"{n:2} {operador} {i:2} = {r:2}")   

tabla(3, "+")
tabla(7, "*")
```

```python
# Funcion internas / if else

def tabla(n, operador="*", cantidad=5, formato="2"):    

    def f(x,y):
        if operador == "+":
            return x + y
        elif operador == "-":
            return x - y
        elif operador == "*":
            return x * y
        elif operador == "/":
            return x / y


    print(f"Tabla del {n}")
    for i in range(1, cantidad + 1):         
        r = f(n, i)
        print(f"{n:2} {operador} {i:2} = {r:{formato}}")   

tabla(3, "/", formato="6.4f")
tabla(7, "*")
```

```python
# Funcion como parametro

def tabla(n, operador="*", f=None, cantidad=5, formato="2"):    

    print(f"Tabla del {n}")
    for i in range(1, cantidad + 1):         
        r = f(n, i)
        print(f"{n:2} {operador} {i:2} = {r:{formato}}")   

def suma(x, y):  return x + y
def resta(x, y): return x - y
def multi(x, y): return x * y
def divi(x, y):  return x / y

tabla(3, "/", f=divi, formato="6.4f")
tabla(7, "*", multi)
```

```python
# Funcion lambda

def tabla(n, operador="*", f=None, cantidad=5, formato="2"):    

    print(f"Tabla del {n}")
    for i in range(1, cantidad + 1):         
        r = f(n, i)
        print(f"{n:2} {operador} {i:2} = {r:{formato}}")   

suma  = lambda x, y: x + y
resta = lambda x, y: x - y
multi = lambda x, y: x * y
divi  = lambda x, y: x / y  # JS > divi = (x, y) => x / y

tabla(3, "/", f=divi, formato="6.4f")
tabla(5, "^", lambda x, y: x ** y, cantidad=10, formato=7)
```

```python
# Diccionar de funciones

suma  = lambda x, y: x + y
resta = lambda x, y: x - y
multi = lambda x, y: x * y
divi  = lambda x, y: x / y  # JS > divi = (x, y) => x / y

#  Usamos un diccionario para elegir la funcion asociada a cada operador
functiones = {
    "+": suma,
    "-": resta,
    "*": multi,
    "/": divi
}

def tabla(n, operador="*", f=None, cantidad=5, formato="2"):    
    if f is None:
        if operador in functiones:
            f = functiones[operador]
        else:
            f = lambda x, y: eval(f"{x}{operador}{y}")

    print(f"Tabla del {n}")
    for i in range(1, cantidad + 1):         
        r = f(n, i)
        print(f"{n:2} {operador} {i:2} = {r:{formato}}")   


tabla(3, "*")   
tabla(5, "^", lambda x, y: x ** y, cantidad=3, formato=4)
```

```python
# Elegir el rango a mostrar

suma  = lambda x, y: x + y
resta = lambda x, y: x - y
multi = lambda x, y: x * y
divi  = lambda x, y: x / y  # JS > divi = (x, y) => x / y

#  Usamos un diccionario para elegir la funcion asociada a cada operador
functiones = {
    "+": suma,
    "-": resta,
    "*": multi,
    "/": divi
}

def tabla(n, operador="*", f=None, cantidad=5, rango=None, formato="2"):    
    
    if f is None:   # Si no le paso una funcion, busco en el diccionario
        if operador in functiones:
            f = functiones[operador]
        else:       # Si no esta en el diccionario, uso eval
            f = lambda x, y: eval(f"{x}{operador}{y}")

    if rango is None:   # Si no le paso un rango, uso el default 
        rango = range(1, cantidad + 1)
    
    print(f"Tabla del {n}")
    for i in rango:         
        print(f"{n:2} {operador} {i:2} = {f(n, i):{formato}}")   


tabla(3, "*", rango = [1, 3, 5, 7, 9])   
```

```python
# Tablas multiples con lista

def tablas(cantidad=10):  
    for n in range(1, cantidad + 1):
        fila = []
        for i in range(1, cantidad + 1):
            fila.append(f"{n/i:6.3f}")
        print(' '.join(fila))


tablas()
```

```python
# Tablas multiples con una lista por comprension

def tablas(cantidad=10):  
    for n in range(1, cantidad + 1):
        fila = [f"{n*i:4}" for i in range(1, cantidad + 1)]
        print(' '.join(fila))

tablas(5)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
df = pd.read_csv('datos_laboratorio_simulados.csv')  # AsegÃºrate de que el archivo estÃ© en la misma carpeta o proporciona la ruta completa

# Convertir los datos a tensores de PyTorch
x_data = torch.tensor(df['x'].values, dtype=torch.float32).unsqueeze(1)
y_data = torch.tensor(df['y'].values, dtype=torch.float32).unsqueeze(1)

# Definir el modelo
class CurveFittingModel(nn.Module):
    def __init__(self):
        super(CurveFittingModel, self).__init__()
        self.hidden1 = nn.Linear(1, 64)  # Primera capa oculta con 64 neuronas
        self.hidden2 = nn.Linear(64, 32) # Segunda capa oculta con 32 neuronas
        self.relu = nn.ReLU()
        self.output = nn.Linear(32, 1)   # Capa de salida

    def forward(self, x):
        x = self.hidden1(x)
        x = self.relu(x)
        x = self.hidden2(x)
        x = self.relu(x)
        x = self.output(x)
        return x
# Instanciar el modelo
model = CurveFittingModel()

# Definir el criterio de pÃ©rdida y el optimizador
criterion = nn.MSELoss()  # Error cuadrÃ¡tico medio
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Entrenar el modelo
num_epochs = 1000
for epoch in range(num_epochs):
    model.train()
    
    # Hacer una predicciÃ³n
    y_pred = model(x_data)
    
    # Calcular la pÃ©rdida
    loss = criterion(y_pred, y_data)
    
    # Backpropagation y optimizaciÃ³n
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    # Imprimir el error cada 100 Ã©pocas
    if (epoch+1) % 100 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

# Cambiar el modelo al modo de evaluaciÃ³n
model.eval()

# Generar predicciones para el conjunto de datos
with torch.no_grad():
    y_pred = model(x_data).numpy()

# Graficar los puntos originales
plt.scatter(df['x'], df['y'], color='blue', label='Datos Originales')

# Graficar la curva ajustada
x_curve = np.linspace(df['x'].min(), df['x'].max(), 100).reshape(-1, 1)
x_curve_tensor = torch.tensor(x_curve, dtype=torch.float32)
y_curve = model(x_curve_tensor).detach().numpy()
plt.plot(x_curve, y_curve, color='red', label='Curva Ajustada')

# Configurar etiquetas y tÃ­tulo
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Ajuste de Curva usando PyTorch')
plt.legend()

# Mostrar el grÃ¡fico
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generar valores de x en un rango, por ejemplo, de -10 a 10
x = np.linspace(-10, 10, 1000)

# Calcular y = x^3 - 3x^2 - 5x - 18
y = x**3 - 3 * x**2 - 5 * x - 18

# Agregar ruido normal con una desviaciÃ³n estÃ¡ndar del 10% del valor de y
ruido = np.random.normal(0, 0.1 * np.std(y), size=y.shape)
y_con_ruido = y + ruido

# Crear un DataFrame con los datos simulados
df_simulado = pd.DataFrame({'x': x, 'y': y_con_ruido})

# Guardar los datos en un archivo CSV
df_simulado.to_csv('datos_laboratorio_simulados.csv', index=False)

# Graficar los datos simulados
plt.scatter(x, y_con_ruido, color='blue', label='Datos Simulados con Ruido')
plt.plot(x, y, color='red', label='FunciÃ³n Original sin Ruido')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Datos Simulados de Laboratorio')
plt.legend()
plt.show()
```
