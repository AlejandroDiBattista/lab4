```python
import pandas as pd

# Crear un DataFrame
data = {
    'Nombre': ['Ana', 'Luis', 'Carlos', 'Marta'],
    'Edad': [23, 45, 34, None],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla']
}
df = pd.DataFrame(data)

# Ver las primeras filas
print("Datos resumen")
print(df.head())

# Rellenar valores nulos
df['Edad'] = df['Edad'].fillna(df['Edad'].mean())

# Filtrar filas
df_filtered = df[df['Edad'] > 30]
print("Datos filtrados")
print(df_filtered)

# Añadir una nueva columna
df['País'] = 'España'

print("Datos original")
print(df)

# Ordenar por Edad

df_sorted = df.sort_values(by='Edad')
print("Datos ordenados")
print(df_sorted)

# Guardar el DataFrame en un archivo CSV
df.to_csv('datos_modificados.csv', index=False)
```

```python
x = np.linspace(-10, 10, 100)
y = 3 * x * x * x * x - 6 * x * x + - 6 * x + 2
yr = y + np.random.normal(0, 10, 100)

fd = pd.DataFrame({'x': x, 'y': y, 'yr': yr})
fd.to_csv('datos_generados.csv', index=False)
plt.scatter(x, yr)
```

```python
ds = pd.read_csv('datos_generados.csv')
plt.plot(ds['x'], ds['y'], c='r')
plt.scatter(ds['x'], ds['yr'])

print(np.sum(ds['yr']))
a=np.polynomial.polynomial.polyfit(ds['x'], ds['yr'], 5)
plt.plot(ds['x'], np.polynomial.polynomial.polyval(ds['x'], a), c='g')
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

# Datos de entrada y salida (sin normalizar)
x = ds.x.values
y = ds.yr.values


# Crear el modelo
model = Sequential()
model.add(Dense(units=16, activation='relu', input_shape=(1,)))
model.add(Dense(units=16, activation='relu'))  # Capa oculta adicional
model.add(Dense(units=1))  # Capa de salida

# Compilar el modelo
model.compile(optimizer='adam', loss='mse')

# Entrenar el modelo
model.fit(x, y, epochs=2500, verbose=1)

# Hacer predicciones
y_pred = model.predict(x)

# Visualizar resultados
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Datos Reales', color='blue')
plt.plot(x, y_pred, label='Curva Ajustada', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

# Datos de entrada y salida (sin normalizar)
x = ds.x.values
y = ds.yr.values

# Crear el modelo
model = Sequential()
model.add(Dense(units=1, input_shape=(1,), use_bias=False, kernel_initializer='zeros'))
model.add(Dense(units=1, use_bias=False, kernel_initializer='zeros'))
model.add(Dense(units=1, use_bias=False, kernel_initializer='zeros'))
model.add(Dense(units=1, use_bias=False, kernel_initializer='zeros'))

# Compilar el modelo
model.compile(optimizer='adam', loss='mse')

# Entrenar el modelo
model.fit(x, y, epochs=500, verbose=1)

# Hacer predicciones
y_pred = model.predict(x)

# Visualizar resultados
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Datos Reales', color='blue')
plt.plot(x, y_pred, label='Curva Ajustada', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
