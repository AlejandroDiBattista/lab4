```python
x = np.linspace(-10, 10, 100)
y = x**3 + 2*x**2 + 3*x + 4
plt.plot(x, y)
plt.show()

# Generar una lista de puntos (x, y)
puntos_x = np.array([-2, -1, 0, 1, 2])
puntos_y = np.array([-15, -1, 1, 3, 15])

# Ajustar un polinomio de grado 3 a los puntos
coeficientes = np.polyfit(puntos_x, puntos_y, 3)
polinomio = np.poly1d(coeficientes)

print("Coeficientes del polinomio:", coeficientes)
print("Polinomio:", polinomio)

# Graficar los puntos y el polinomio propuesto
plt.scatter(puntos_x, puntos_y, color='red', label='Puntos')
plt.plot(x, polinomio(x), label='Polinomio ajustado')
plt.legend()
plt.show()
```

```python
x = np.linspace(-10, 10, 100)

y = 3 * x ** 3 - 2 * x ** 2
yr = y + np.random.normal(0, 200, 100)
# plt.plot(x, y, linewidth=4, color='red')
plt.scatter(x, yr, color='blue')
plt.grid(True)
plt.minorticks_on()
plt.grid(which='both', linestyle=':', linewidth='0.5')
plt.grid(which='major', linestyle='-', linewidth='1.0', color='gray')

coeficientes_yr = np.polyfit(x, yr, 3)
polinomio_yr = np.poly1d(coeficientes_yr)

print("Coeficientes del polinomio para yr:", coeficientes_yr)
print("Polinomio para yr:", polinomio_yr)

plt.plot(x, polinomio_yr(x), label='Polinomio ajustado para yr', color='green')
plt.legend()
plt.show()
```
