## For
En Python no existe el `for (let i=0; i<10; i++)`.
El `for` en Python es un `foreach`.

Permite recorrer una secuencia de elementos (listas, tuplas, diccionarios, strings).

GPT: ¿Cuál es la sintaxis del `for`? Dame ejemplos de uso.

```python
# c       => for (int i=1; i<11; i++) { ... }
# c++     => for (int i=1; i<11; i++) { ... }
# c#      => for (int i=1; i<11; i++) { ... }
# java    => for (int i=1; i<11; i++) { ... }
# JS      => for (let i=1; i<11; i++) { ... }
# dart    => for (int i=1; i<11; i++) { ... }
# swift   => for i in 1...10 { ... }
# ruby    => for i in 1..10 do ; end 
# basic   => for i = 1 to 10 : next i
# fortran => do i = 1, 10, 1
# pascal  => for i := 1 to 10 do
# perl    => for (my $i=1; $i<11; $i++) { ... }
# php     => for ($i=1; $i<11; $i++) { ... }
# python  => for i in range(1,11):
# cobol   => perform varying i from 1 by 1 until i > 10

n = 8
for i in range(5,0,-1):
    print(f"{n:2} * {i:2} = {n*i:3}")
```

# Lista con comprensión.

Para crear listas existe una forma especializada llamada listas por comprensión que pone la fórmula que deben cumplir los elementos de la lista.

Consiste en un `for` para recorrer una secuencia y opcionalmente un `if` para filtrar elementos.

GPT: ¿Qué son las listas por comprensión? Dame tres ejemplos y sus equivalentes hechos con `for`.

```python
lista =[]
for i in range(1,10):
    if i % 2 == 0:
        lista.append(i*i)
        
lista2 = [i*i for i in range(1,10) 
                if i % 2 == 0]
print(lista)
print(lista2)
```

Las listas pueden ser combinadas.

```python
a = [[x*y for x in range(1,3)] for y in range(1,4)]

a = []
for y in range(1, 4):
    row = []
    for x in range(1, 3):
        row.append(x * y)
    a.append(row)
a   
# JS let a = Array.from({ length: 2 }, (_, y) => Array.from({ length: 2 }, (_, x) => (x + 1) * (y + 1))

```

# Definición de funciones

Las funciones en Python son equivalentes a las de JavaScript, pero son más flexibles en su uso.

Un elemento distintivo es que no se usan `{}` sino indentación para definir cuáles son las sentencias dentro de la definición de la función.

Los parámetros pueden tener valores por defecto (lo que los hace opcionales).

La llamada a una función se puede hacer de forma posicional (los parámetros se asignan según la posición) o por nombre.

GPT: ¿Cuál es la forma de pasar parámetros a una función? ¿Qué es la asignación posicional? ¿Qué otra forma hay de asignar?

# Librerías poderosas

NumPy es una librería para manejar arrays y hacer operaciones sobre ellos.

GPT: ¿Qué es NumPy? ¿Para qué se usa? Dame ejemplos.

```python
import numpy as np

def cuadratica(x):
    return 2 * x * x + 3 * x - 5

a = np.arange(1,10,2) # [1,3,5,7,9]
print(a)

b = a + 3  # Suma 3 a cada elemento
print(b)

b = a * 3 # Multiplica por 3 a cada elemento
print(b)

b = a * a # Multiplica elemento a elemento
print(b)


print(cuadratica(a))    # aplica la función cuadratica a cada elemento de a
```

Incluso se puede hacer graficos 3D

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la función para calcular Z y graficar en 3D
def graficar_3d():
    # Crear los valores para X y Y en el rango de -5 a 5
    x = np.arange(-5, 6, 1)
    y = np.arange(-5, 6, 1)
    
    # Crear la malla de valores X, Y
    X, Y = np.meshgrid(x, y)
    
    # Calcular Z como el producto de X e Y dividido por 5
    Z = (X * Y * X) / 5
    
    # Crear la figura y el gráfico 3D
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    # Graficar la superficie
    ax.plot_surface(X, Y, Z, cmap='viridis')
    
    # Etiquetas de los ejes
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    
    # Mostrar el gráfico
    plt.show()

# Llamar a la función para graficar
graficar_3d()
```
