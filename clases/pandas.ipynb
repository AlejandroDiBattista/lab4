# Pandas
Pandas es una biblioteca de Python utilizada para la manipulación y análisis de datos. Proporciona estructuras de datos como DataFrames y Series, que facilitan el manejo de datos tabulares y de series temporales.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi,np.pi,10000)
ys = np.sin(x)
cs = 3 * np.cos(2 * x)

plt.plot(x,ys)
plt.plot(x,cs)
l = pd.DataFrame({
    'x':x,
    'y=sin(x)':ys,
    'y=cos(x)':cs, 
})

l.to_csv('sin_cos.csv',index=False)
l.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

import pandas as pd

df = pd.read_csv('sin_cos.csv')
x = df['x']
y = df['y=sin(x)']

yr = y + np.random.normal(0,0.1,len(y))

plt.ylim(-2, 2)  # Establecer los límites del eje y de -2 a 2
plt.plot(x,y,'r', linewidth=5,alpha=0.5)
plt.scatter(x, yr, s=2)  # Ajusta el valor de 's' según el tamaño deseado

# Agregar la grilla principal cada 1 unidad
plt.grid(True)
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(1))

# Agregar la subgrilla cada 0.1 unidades
plt.minorticks_on()
plt.gca().xaxis.set_minor_locator(ticker.MultipleLocator(0.1))
plt.gca().yaxis.set_minor_locator(ticker.MultipleLocator(0.1))
plt.grid(True, which='minor', linestyle=':', linewidth='0.5')


```
